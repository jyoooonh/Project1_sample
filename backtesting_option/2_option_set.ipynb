## CTB0915 콜옵션 23년 11월물 9시15분 현재가

```python
root_directory = '../option_price/201TB/'
CTB0915 = pd.DataFrame()
for root, dirs, files in os.walk(root_directory):    
    for file in files:
        file_path = os.path.join(root, file)        
        # data = pd.read_csv(file_path)
        csv_file_name = os.path.basename(file_path)
        data = pd.read_csv(file_path)
        data['현재가'] = data['현재가'].abs()        
        csv_file_name = csv_file_name.split('.')[0].split('TB')[1]
        data['체결시간'] = pd.to_datetime(data['체결시간'], format="%Y%m%d%H%M%S")
        data = data[['체결시간', '현재가']]
        data.rename(columns={'현재가': csv_file_name, '체결시간': 'date'}, inplace=True)
        if CTB0915.empty:
            CTB0915 = data
        else:
            CTB0915 = pd.merge(CTB0915, data, on='date', how='outer')
CTB0915.sort_values(by='date', ascending=False, inplace=True)
# CTB0915.fillna(method='bfill', inplace=True)
CTB0915 = CTB0915[CTB0915['date'].dt.time == pd.to_datetime('09:15:00').time()] # 09시30분 데이터만 추출
CTB0915['date'] = CTB0915['date'].dt.strftime("%Y-%m-%d")
CTB0915.set_index('date', inplace=True)
```

# 거래일 기간 조정

```python
import datetime

def find_second_thursdays(year, month):
    current_month_second_thursday = datetime.date(year, month, 1)
    current_month_second_thursday += datetime.timedelta(days=(3 - current_month_second_thursday.weekday() + 7) % 7)
    current_month_second_thursday += datetime.timedelta(weeks=1)

    if month == 1:
        previous_month = 12
        previous_year = year - 1
    else:
        previous_month = month - 1
        previous_year = year

    previous_month_second_thursday = datetime.date(previous_year, previous_month, 1)
    previous_month_second_thursday += datetime.timedelta(days=(3 - previous_month_second_thursday.weekday() + 7) % 7)
    previous_month_second_thursday += datetime.timedelta(weeks=1)

    return current_month_second_thursday, previous_month_second_thursday

# 사용 예제
year = 2023
month = 11

current_month_second_thursday, previous_month_second_thursday = find_second_thursdays(year, month)
current_month_second_thursday = str(current_month_second_thursday)
previous_month_second_thursday = str(previous_month_second_thursday)
print(f"{year}년 {month}월의 두 번째 목요일: {current_month_second_thursday}")
print(f"{year}년 {month}월의 전월 두 번째 목요일: {previous_month_second_thursday}")
```
