```python
import pandas as pd
```

```python
kospi = pd.read_csv('../dataset/kospi_30min.csv', index_col=0)
kospi
```

```python
kospi.drop(['대업종구분', '소업종구분', '종목정보', '전일종가'], axis=1, inplace=True)
```

```python
kospi['체결시간'] = kospi['체결시간'].astype(str)
```

```python
kospi = kospi[kospi['체결시간'].str.contains('090000|093000')]
```

```python
kospi = kospi.reset_index(drop=True)
```

```python
kospi = kospi[['현재가', '체결시간', '고가', '저가']]
```

```python
kospi
```

```python
kospi['현재가'] = abs(kospi['현재가'])
kospi['고가'] = abs(kospi['고가'])
kospi['저가'] = abs(kospi['저가'])
```

```python
kospi['현재가'] = kospi['현재가'] / 100
kospi['고가'] = kospi['고가'] / 100
kospi['저가'] = kospi['저가'] / 100
```

```python
kospi
```

```python
date = []
rate_of_change = []
fluc_range = []

for i in range(0, len(kospi), 2):
	date.append(kospi['체결시간'][i][:8])
	change_result = (kospi['현재가'][i] - kospi['현재가'][i + 1]) / kospi['현재가'][i + 1] * 100

	if kospi['저가'][i] > kospi['저가'][i + 1]:
		min = kospi['저가'][i + 1]
	else:
		min = kospi['저가'][i]

	if kospi['고가'][i] > kospi['고가'][i + 1]:
		max = kospi['고가'][i]
	else:
		max = kospi['고가'][i + 1]

	fluc_result = max - min

	fluc_result = round(fluc_result, 3)
	change_result = round(change_result, 3)

	rate_of_change.append(change_result)
	fluc_range.append(fluc_result)
```

```python
kospi = pd.DataFrame(zip(date, rate_of_change, fluc_range), columns = ['date', 'kospi_p_change(%)', 'kospi_p_range'])
```

```python
kospi
```

```python
kospi = kospi[5:59]
```

```python
kospi = kospi[::-1].reset_index(drop=True)
```

```python
kospi['date'] = pd.to_datetime(kospi['date'])
```
