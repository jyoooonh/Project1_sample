```python
import pandas as pd
```

```python
pc_ratio = pd.read_csv('../dataset/put_call_ratio.csv', encoding='cp949')
pc_ratio
```

```python
pc_ratio.info()
```

```python
pc_ratio = pc_ratio.rename(columns={'일자':'date'})
```

```python
pc_ratio.date = pd.to_datetime(pc_ratio.date)
```

```python
pc_ratio['P/C Ratio'] = pc_ratio['P/C Ratio'].shift(-1)
```

```python
pc_ratio['put_shift'] =  pc_ratio['거래량_PUT'].shift(1)
pc_ratio['call_shift'] =  pc_ratio['거래량_CALL'].shift(1)
```

```python
pc_ratio['put_volume_fluc'] = round(((pc_ratio['put_shift'] - pc_ratio['거래량_PUT']) / pc_ratio['거래량_PUT']) * 100, 3)
pc_ratio['call_volume_fluc'] = round(((pc_ratio['call_shift'] - pc_ratio['거래량_CALL']) / pc_ratio['거래량_CALL']) * 100, 3)
pc_ratio
```

```python
pc_ratio = pc_ratio[['date', 'P/C Ratio', 'put_volume_fluc', 'call_volume_fluc']]
```

```python
pc_ratio['put_volume_fluc'] = pc_ratio['put_volume_fluc'].shift(-2)
pc_ratio['call_volume_fluc'] = pc_ratio['call_volume_fluc'].shift(-2)
pc_ratio
```

```python
pc_ratio = pc_ratio[4:58]
```

```python
pc_ratio = pc_ratio[::-1].reset_index(drop=True)
```
