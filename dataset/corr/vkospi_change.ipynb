```python
import pandas as pd
```

```python
vkospi_half = pd.read_csv('../data/VKOSPI.csv')
vkospi_half
```

```python
vkospi = pd.read_csv('../test_feature/VKOSPI.csv')
vkospi[:2]
```

```python
vkospi = pd.concat([vkospi[:2], vkospi_half], axis=0)
```

```python
vkospi
```

```python
new_row = {'날짜':['2023- 11- 03']}
new_row = pd.DataFrame(new_row)
new_row.index = [0]
vkospi = pd.concat([new_row, vkospi], ignore_index=True)
vkospi
```

```python
vkospi[['종가', '시가']] = vkospi[['종가', '시가']].shift(-1)
```

```python
vkospi
```

```python
vkospi = vkospi[::-1]
```

```python
vkospi
```

```python
vkospi = vkospi.reset_index(drop=True)
```

```python
vkospi
```

```python
vkospi['VKOSPI 당일 변동율'] = round((vkospi['종가'] - vkospi['시가']) / vkospi['시가'] * 100, 3)
```

```python
vkospi
```

```python
vkospi[206:]
```

```python
vkospi['날짜'] = pd.to_datetime(vkospi['날짜'])
```
