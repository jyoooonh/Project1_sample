```python
import pandas as pd 
import matplotlib.pyplot as plt

import seaborn as sns
```

```python
data_two_month_corr_data = pd.read_csv('./two_month_corr_data.csv', index_col=0)
data_two_month_corr_data.head(3)
```

```python
quater_corr_data = pd.read_csv('./half_corr_data.csv', index_col=0)
quater_corr_data
```

```python
# data_two_month_corr_data.set_index('날짜', inplace=True)

# 상관관계를 계산할 특정 열을 정의합니다.
x = [
    "시장베이시스",
    "이론베이시스",
    "괴리율",
    "원위안 야간변동율",
    "원엔 야간변동율",
    "KOSPI 전날변동율",
    "KOSPI 야간변동율",
    "KS200 전날변동율",
    "KS200 야간변동율",
    "NAS 야간변동율",
    "NAS 당일변동율",
    "NAS 당일변화량",
    "P/C Ratio",
    "CALL_vol_change(%)",
    "PUT_vol_change(%)",
    "CALL_vol_fluc",
    "PUT_vol_fluc",
    "S&P 야간변동율",
    "S&P 당일변동율",
    "S&P 당일변화량",
    "원달러 야간변동율",
    "VIX 전날변동율",
    "VIX 당일변동율",
    "VIX 당일변화량",
    "JNIV 종가변동율",
    "JNIV 전날변동율",
    "JNIV 전날변화량",
    "CD 전날변동율",
    "CD 전날변화량",
    "KOSPI 오전변동율",
    "KOSPI 오전변화량",
    "KS200 오전변동율",
    "KS200 오전변화량",
]

# 상관관계를 계산합니다.
correlations = data_two_month_corr_data[x].corrwith(
    data_two_month_corr_data["VKOSPI 당일 변동율"]
)

# 상관관계를 내림차순으로 정렬합니다.
sorted_correlations = correlations.sort_values(ascending=False)

# 상관관계를 시각화합니다.
plt.figure(figsize=(10, 8))
sns.barplot(x=sorted_correlations.values, y=sorted_correlations.index)
plt.title("Correlation with VKOSPI Daily Change Rate")
plt.xlabel("Correlation Coefficient")
plt.ylabel("Variables")
plt.tight_layout()  # 레이아웃을 조정하여 레이블이 잘리는 것을 방지합니다.
plt.show()

plt.savefig("./output.png", bbox_inches="tight")

plt.rcParams["font.family"] = "AppleGothic"
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


# 상관관계를 계산할 특정 열을 정의합니다.
x = [
    "시장베이시스",
    "이론베이시스",
    "괴리율",
    "원위안 야간변동율",
    "원엔 야간변동율",
    "KOSPI 전날변동율",
    "KOSPI 야간변동율",
    "KS200 전날변동율",
    "KS200 야간변동율",
    "NAS 야간변동율",
    "NAS 당일변동율",
    "NAS 당일변화량",
    "P/C Ratio",
    "CALL_vol_change(%)",
    "PUT_vol_change(%)",
    "CALL_vol_fluc",
    "PUT_vol_fluc",
    "S&P 야간변동율",
    "S&P 당일변동율",
    "S&P 당일변화량",
    "원달러 야간변동율",
    "VIX 전날변동율",
    "VIX 당일변동율",
    "VIX 당일변화량",
    "JNIV 종가변동율",
    "JNIV 전날변동율",
    "JNIV 전날변화량",
    "CD 전날변동율",
    "CD 전날변화량",
]

# 상관관계를 계산합니다.
correlations = quater_corr_data[x].corrwith(quater_corr_data["VKOSPI 당일 변동율"])

# 상관관계를 내림차순으로 정렬합니다.
sorted_correlations = correlations.sort_values(ascending=False)

# 폰트 설정
plt.rcParams["font.family"] = "AppleGothic"  # 맥북의 기본 한글 폰트인 AppleGothic을 사용합니다.
plt.rcParams["axes.unicode_minus"] = False  # 마이너스 부호를 올바르게 표시합니다.

# 그래프의 크기를 크게 설정합니다.
plt.figure(figsize=(12, 10))

# 상관관계를 시각화합니다.
sns.barplot(x=sorted_correlations.values, y=sorted_correlations.index)
plt.title("VKOSPI 당일 변동율과의 상관관계")
plt.xlabel("상관계수")
plt.ylabel("변수")
plt.xticks(rotation=45)  # x 축 레이블을 45도 회전시켜 표시
plt.tight_layout()  # 레이아웃 조정

# 이미지 파일로 저장할 때 글자가 잘리지 않도록 설정
#plt.savefig("./output.png", bbox_inches="tight")

# 그래프를 표시
plt.show()
```
