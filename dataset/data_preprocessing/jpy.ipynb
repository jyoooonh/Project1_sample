```python
import pandas as pd
```

```python
jpy = pd.read_csv('../data/JPY_KRW.csv', index_col=0)
jpy
```

```python
jpy = jpy[::-1]
```

```python
jpy = jpy.reset_index()
```

```python
jpy = jpy[['날짜', '종가', '시가']]
```

```python
jpy
```

```python
jpy['종가'] = jpy['종가'].shift(1)
```

```python
jpy['야간 변동률'] = round((jpy['시가'] - jpy['종가']) / jpy['종가'] * 100, 3)
```

```python
jpy = jpy.dropna()
```

```python
jpy = jpy[1:266]
```

```python
jpy = jpy.reset_index(drop=True)
```

```python
jpy['날짜'] = pd.to_datetime(jpy['날짜'])
```

```python
jpy[120:]
```

```python
jpy[125:255].reset_index(drop=True)[['날짜', '야간 변동률']].to_csv('../half_clean_feature/jpy_half.csv')
```

```python
jpy[170:255].reset_index(drop=True)[['날짜', '야간 변동률']].to_csv('../quater_clean_feature/jpy_quater.csv')
```
