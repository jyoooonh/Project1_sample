```python
import pandas as pd
```

```python
cny = pd.read_csv('../test_feature/CNY_KRW.csv')
cny
```

```python
cny = cny[::-1]
```

```python
cny = cny.reset_index()
```

```python
cny = cny[['날짜', '종가', '시가']]
```

```python
cny['종가'] = cny['종가'].shift(1)
```

```python
cny['야간 변동률'] = round((cny['시가'] - cny['종가']) / cny['종가'] * 100, 3)
```

```python
cny = cny.dropna()
```

```python
cny = cny[4:]
```

```python
cny = cny.reset_index(drop=True)
```

```python
cny['날짜'] = pd.to_datetime(cny['날짜'])
```

```python
cny
```

```python
cny.reset_index(drop=True)[['날짜', '야간 변동률']].to_csv("../test_feature/clean_feature/cny.csv")
```
