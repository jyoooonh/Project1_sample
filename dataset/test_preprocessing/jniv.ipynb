```python
import pandas as pd
```

```python
jniv = pd.read_csv('../test_feature/JNIV.csv')
jniv
```

```python
new_row = {'날짜':['2023- 11- 03']}
new_row = pd.DataFrame(new_row)
new_row.index = [0]
jniv = pd.concat([new_row, jniv], ignore_index=True)
jniv
```

```python
jniv = jniv[::-1]
jniv.info()
```

```python
jniv = jniv.reset_index(drop=True)
```

```python
jniv = jniv[['날짜', '종가', '시가', '고가', '저가', '변동 %']]
```

```python
jniv['변동 %'] = jniv['변동 %'].astype(str)
jniv["변동 %"] = jniv["변동 %"].str.rstrip('%')
jniv["변동 %"] = jniv["변동 %"].astype(float)
```

```python
jniv['변동 %'] = jniv['변동 %'].shift(2)
jniv['종가'] = jniv['종가'].shift(1)
jniv
```

```python
jniv['전날 변동율'] = round((jniv['종가'] - jniv['시가']) / jniv['시가'] * 100, 3)
jniv['전날 변화량'] = round(jniv['고가'] - jniv['저가'], 3)
```

```python
jniv['전날 변동율'] = jniv['전날 변동율'].shift(1)
jniv['전날 변화량'] = jniv['전날 변화량'].shift(1)
```

```python
jniv = jniv[4:]
```

```python
jniv = jniv.reset_index(drop=True)
```

```python
jniv['날짜'] = pd.to_datetime(jniv['날짜'])
```

```python
jniv
```

```python
jniv.reset_index(drop=True)[['날짜', '변동 %', '전날 변동율', '전날 변화량']].to_csv('../test_feature/clean_feature/jniv.csv')
```
