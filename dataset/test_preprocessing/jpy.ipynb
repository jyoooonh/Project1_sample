```python
import pandas as pd
```

```python
jpy = pd.read_csv("../test_feature/JPY_KRW.csv")
jpy
```

```python
jpy = jpy[::-1]
```

```python
jpy = jpy.reset_index()
```

```python
jpy = jpy[['날짜', '종가', '시가']]
```

```python
jpy
```

```python
jpy['종가'] = jpy['종가'].shift(1)
```

```python
jpy['야간 변동률'] = round((jpy['시가'] - jpy['종가']) / jpy['종가'] * 100, 3)
```

```python
jpy = jpy.dropna()
```

```python
jpy
```

```python
jpy = jpy[4:]
```

```python
jpy = jpy.reset_index(drop=True)
```

```python
jpy['날짜'] = pd.to_datetime(jpy['날짜'])
```

```python
jpy[['날짜', '야간 변동률']].reset_index(drop=True).to_csv('../test_feature/clean_feature/jpy.csv')
```
