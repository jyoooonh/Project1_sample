```python
import pandas as pd
```

```python
ks200_mor = pd.read_csv('../test_feature/ks200_mor.csv')
ks200_mor
```

```python
ks200_mor.drop(['대업종구분', '소업종구분', '종목정보', '전일종가'], axis=1, inplace=True)
```

```python
ks200_mor['체결시간'] = ks200_mor['체결시간'].astype(str)
```

```python
ks200_mor = ks200_mor[ks200_mor['체결시간'].str.contains('090000|093000')]
```

```python
ks200_mor
```

```python
ks200_mor = ks200_mor.reset_index(drop=True)
```

```python
ks200_mor = ks200_mor[['현재가', '체결시간', '고가', '저가']]
```

```python
ks200_mor['현재가'] = abs(ks200_mor['현재가'])
ks200_mor['고가'] = abs(ks200_mor['고가'])
ks200_mor['저가'] = abs(ks200_mor['저가'])
```

```python
ks200_mor['현재가'] = ks200_mor['현재가'] / 100
ks200_mor['고가'] = ks200_mor['고가'] / 100
ks200_mor['저가'] = ks200_mor['저가'] / 100
```

```python
ks200_mor
```

```python
date = []
rate_of_change = []
fluc_range = []

for i in range(0, len(ks200_mor), 2):
	date.append(ks200_mor['체결시간'][i][:8])
	change_result = (ks200_mor['현재가'][i] - ks200_mor['현재가'][i + 1]) / ks200_mor['현재가'][i + 1] * 100

	if ks200_mor['저가'][i] > ks200_mor['저가'][i + 1]:
		min = ks200_mor['저가'][i + 1]
	else:
		min = ks200_mor['저가'][i]

	if ks200_mor['고가'][i] > ks200_mor['고가'][i + 1]:
		max = ks200_mor['고가'][i]
	else:
		max = ks200_mor['고가'][i + 1]

	fluc_result = max - min

	fluc_result = round(fluc_result, 3)
	change_result = round(change_result, 3)

	rate_of_change.append(change_result)
	fluc_range.append(fluc_result)
```

```python
ks200_mor = pd.DataFrame(zip(date, rate_of_change, fluc_range), columns = ['date', 'KS200 오전변동율', 'KS200 오전변화량'])
```

```python
ks200_mor
```

```python
ks200_mor = ks200_mor[:19]
```

```python
ks200_mor = ks200_mor[::-1].reset_index(drop=True)
```

```python
ks200_mor['date'] = pd.to_datetime(ks200_mor['date'])
```
