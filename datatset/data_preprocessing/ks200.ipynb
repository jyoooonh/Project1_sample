```python
import pandas as pd
```

```python
ks200 = pd.read_csv('../data/KOSPI200.csv', index_col=0)
ks200
```

```python
ks200 = ks200[::-1]
ks200
```

```python
ks200 = ks200.reset_index()
```

```python
ks200.head(3)
```

```python
ks200 = ks200[['날짜', '종가', '시가', '변동 %']]
```

```python
ks200['변동 %'] = ks200['변동 %'].astype(str)
ks200["변동 %"] = ks200["변동 %"].str.rstrip('%')
ks200["변동 %"] = ks200["변동 %"].astype(float)
```

```python
ks200['변동 %'] = ks200['변동 %'].shift(1)
ks200
```

```python
ks200 = ks200.dropna()
```

```python
ks200['종가'] = ks200['종가'].shift(1)
```

```python
ks200['야간변화율'] = round(((ks200['시가'] - ks200['종가']) / ks200['종가']) * 100, 3)
```

```python
ks200 = ks200.dropna()
ks200.info()
```

```python
ks200['날짜'] = pd.to_datetime(ks200['날짜'])
```

```python
ks200 = ks200[:251]
```

```python
ks200 = ks200.reset_index(drop=True)
```

```python
ks200
```

```python
ks200[['날짜', '변동 %', '야간변화율']].to_csv('../year_clean_feature/ks200.csv')
```
