```python
import pandas as pd
```

```python
vkospi = pd.read_csv('../data/VKOSPI.csv', index_col=0)
```

```python
vkospi
```

```python
vkospi = vkospi[::-1]
```

```python
vkospi = vkospi.reset_index()
```

```python
label = []

for i in range(len(vkospi['날짜'])):
	if vkospi['시가'][i] > vkospi['종가'][i]:
		result = 0
		label.append(result)
	else:
		result = 1
		label.append(result)

print(label)
```

```python
vkospi
```

```python
label_df = pd.DataFrame(label, columns=['VKOSPI_Label'])
```

```python
vkospi = pd.concat([vkospi, label_df], axis=1)
```

```python
vkospi
```

```python
vkospi = vkospi[4:255]
```

```python
vkospi
```

```python
vkospi['날짜'] = pd.to_datetime(vkospi['날짜'])
```

```python
vkospi[['날짜', 'VKOSPI_Label']].reset_index(drop=True).to_csv("../year_clean_feature/vkospi_year.csv")
```

```python
vkospi[116:][['날짜', 'VKOSPI_Label']].reset_index(drop=True).to_csv("../half_clean_feature/vkospi_half.csv")
```
