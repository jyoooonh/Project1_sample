```python
import pandas as pd
```

```python
strike_price = pd.read_csv('./strike_price.csv')
strike_price = strike_price[:31]
strike_price = strike_price[::-1]
```

```python
df = pd.DataFrame()
for date in strike_price['일자']:
	data = pd.read_csv(f'./option_day_strike_price/{date}_clean.csv', index_col=0)
	data = pd.pivot_table(data, index=['거래일자', '풋콜구분', '시각'], columns='행사가', values=['종가'])
	
	data = data.reset_index()
	data = data.fillna(method='ffill')

	buy_data = data[(data['시각'] == 915)]
	#sell_data = data[(data['시각'] == 1515) | (data['시각'] == 1530)]
	sell_data = data[(data['시각'] == 1515)]

	df = pd.concat([df, buy_data, sell_data])

df = pd.pivot_table(df, index=['거래일자', '풋콜구분', '시각'], values=['종가'])
#df.to_csv('./buy_sell_1515.csv')
```

```python
import pandas as pd

# 원본 데이터프레임을 가정합니다.
# 데이터를 로드하는 코드가 필요하며 데이터프레임 이름은 df로 가정합니다.

# 새로운 데이터프레임을 생성합니다.
profit_data = pd.DataFrame(columns=['거래일자', '행사가', '풋콜구분', '수익'])

# '행사가' 값을 columns에서 가져오기
strike_prices = df.columns.levels[1]

# 모든 거래일자, 행사가, 풋콜구분 조합에 대해 반복합니다.
for date in df.index.get_level_values('거래일자').unique():
    for strike_price in strike_prices:
        for option_type in df.index.get_level_values('풋콜구분').unique():
            # 특정 날짜(date)에 대한 데이터를 가져옵니다.
            selected_date_data = df[df.index.get_level_values('거래일자') == date]

            # 9시에 콜/풋 옵션을 사는 가격을 가져옵니다.
            try:
                buy_price = selected_date_data['종가', strike_price][date][option_type][915]
            except KeyError:
                buy_price = None

            # 15시에 콜/풋 옵션을 팔아서 얻는 가격을 가져옵니다.
            try:
                sell_price = selected_date_data['종가', strike_price][date][option_type][1515]
            except KeyError:
                sell_price = None

            # NaN 값이 있는 경우에 대비하여 수익 계산
            if buy_price is not None and sell_price is not None:
                profit = sell_price - buy_price
            else:
                profit = None

            # 결과를 새로운 데이터프레임에 추가
            if profit is not None:
                profit_data = pd.concat([profit_data, pd.DataFrame({
                    '거래일자': [date],
                    '행사가': [strike_price],
                    '풋콜구분': [option_type],
                    '수익': [profit]
                })], ignore_index=True)

# 결과 데이터프레임을 출력하거나 저장할 수 있습니다.
print(profit_data)
```

```python
profit_data.to_csv("./profit.csv")
```
