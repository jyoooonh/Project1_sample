```python
strike_price = pd.read_csv('../strike_price.csv')
strike_price = strike_price[7:45]
strike_price = strike_price[::-1]
strike_price = strike_price.drop(index=23)
strike_price = strike_price.reset_index(drop=True)
strike_price = strike_price[['일자', 'S-4', 'S-3', 'S-2', 'S-1', '기준행사가', 'S+1', 'S+2', 'S+3', 'S+4']]
strike_price.columns = ['일자', 'S-4', 'S-3', 'S-2', 'S-1', 'S', 'S+1', 'S+2', 'S+3', 'S+4']
strike_price.head(5)
```

```python
# S+1 CALL / S-1 PUT 스트랭글
strangle = pd.DataFrame(columns=["일자"])

for i in range(0, len(strike_price)):
    j = i * 2
    # column
    call = strike_price["S+1"][i]  # Call
    put = strike_price["S-1"][i]  # PUT

    # 날짜
    date = df["거래일자"][j]
    price = round(df[str(call)][j] + df[str(put)][j + 1], 3)

    data = pd.DataFrame({"일자": date, "S1": price}, index=[0])
    strangle = pd.concat([strangle, data])
strangle_S1 = strangle.reset_index(drop=True)
strangle_S1
buy_date_S1 = strangle_S1.loc[strangle_S1["S1"] > 0].일자
```

```python
profit = pd.merge(strangle_S1, strangle_S2, on='일자')
```

```python
profit = pd.merge(profit, strangle_S3, on="일자")
```

```python
profit = pd.merge(profit, strangle_S4, on='일자')
```

```python
# #df = pd.merge(
#     df_profit, vkospi["VKOSPI_Label"], left_index=True, right_index=True, how="right"
# )
profit = pd.merge(
    profit,
    vkospi['VKOSPI_Label'],
    left_index=True, right_index=True, how='right'
)
```

```python
profit = profit.fillna(0)
```

```python
profit_table = pd.DataFrame(
    columns=["S1", "S2", "S3", "S4"],
    index=["양매도", "양매수"],
)
```

```python
sell_total = 0
buy_total = 0
for index, i in enumerate(profit["VKOSPI_Label"]):
    if i == 1:
        buy_value = round(profit["S1"][index], 2)
        buy_total += buy_value
    else:
        sell_value = round(profit["S1"][index], 2)
        sell_total += sell_value
print(f"양매도 전략 : {sell_total}")
print(f"양매수 전략 : {buy_total}")

profit_table.at["양매도", "S1"] = sell_total
profit_table.at["양매수", "S1"] = buy_total
```
