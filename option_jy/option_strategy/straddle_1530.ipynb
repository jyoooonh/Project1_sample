```python
import pandas as pd
```

```python
df = pd.read_csv('../pivot_profit_1530.csv')
df
```

```python
vkospi = pd.read_csv('../vkospi.csv', index_col=0)
vkospi = vkospi.drop(index=23)
vkospi = vkospi.reset_index(drop=True)
vkospi
```

```python
df_profit = df.groupby(['거래일자']).sum().reset_index()
df_profit
```

```python
strike_price = pd.read_csv('../strike_price.csv')
strike_price = strike_price[7:45]
strike_price = strike_price[::-1]
strike_price = strike_price.drop(index=23)
strike_price = strike_price.reset_index(drop=True)
strike_price
```

```python
df = pd.merge(df_profit, vkospi['VKOSPI_Label'], left_index=True, right_index=True, how='right')
df
```

```python
strike_price
```

```python
profit = pd.DataFrame(columns=['S-4', 'S-3', 'S-2', 'S-1', 'S', 'S+1', 'S+2', 'S+3', 'S+4'], index=['양매도', '양매수'])
```

```python
# S-4 스트래들 수익
sell_total = 0
buy_total = 0
for index, i in enumerate(strike_price['S-4']):
	if df['VKOSPI_Label'][index] == 1:
		buy_value = round(df[str(i)][index], 2)
		buy_total += buy_value
	else:
		sell_value = round(df[str(i)][index], 2)
		sell_total += sell_value
print(f'양매도 전략 : {sell_total}')
print(f'양매수 전략 : {buy_total}')

profit.at['양매도', 'S-4'] = sell_total
profit.at['양매수', 'S-4'] = buy_total
```
