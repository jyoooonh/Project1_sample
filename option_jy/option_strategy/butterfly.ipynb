```python
import pandas as pd
```

```python
df = pd.read_csv("../pivot_profit_1515.csv")
df_call = df.loc[df["풋콜구분"] == "CALL"].reset_index(drop=True)
df_put = df.loc[df["풋콜구분"] == "PUT"].reset_index(drop=True)
df
```

```python
vkospi = pd.read_csv("./vkospi.csv", index_col=0)
vkospi = vkospi.drop(index=23)
vkospi = vkospi.reset_index(drop=True)
vkospi
```

```python
strike_price = pd.read_csv("../strike_price.csv")
strike_price = strike_price[7:45]
strike_price = strike_price[::-1]
strike_price = strike_price.drop(index=23)
strike_price = strike_price.reset_index(drop=True)
strike_price = strike_price[
    ["일자", "S-4", "S-3", "S-2", "S-1", "기준행사가", "S+1", "S+2", "S+3", "S+4"]
]
strike_price.columns = [
    "일자",
    "S-4",
    "S-3",
    "S-2",
    "S-1",
    "S",
    "S+1",
    "S+2",
    "S+3",
    "S+4",
]
strike_price.head(5)
```

```python
# S-1 / S+1 -> 매수 / S 매도 버터플라이
butterfly_buy_1 = pd.DataFrame(columns=["일자"])
butterfly_sell_1 = pd.DataFrame(columns=["일자"])

for i in range(0, len(strike_price)):
    left = strike_price["S-1"][i]
    right = strike_price["S+1"][i]
    center = strike_price["S"][i]

    date = strike_price["일자"][i]

    # [CALL] left / right 매수 -> center 매도 (버터플라이 매수 call(변동성 작아질때))
    call_buy_price = round(
        df_call[str(left)][i] + df_call[str(right)][i] - (2 * df_call[str(center)][i]),
        4,
    )
    # [CALL] left / right 매도 -> center 매수 (버터플라이 매도 call(변동성 커질때))
    call_sell_price = -round(
        df_call[str(left)][i] + df_call[str(right)][i] - (2 * df_call[str(center)][i]),
        4,
    )
    # [PUT] left / right 매수 -> center 매도 (버터플라이 매수 put(변동성 작아질때))
    put_buy_price = round(
        df_put[str(left)][i] + df_put[str(right)][i] - (2 * df_put[str(center)][i]), 4
    )
    # [PUT] left / right 매도 -> center 매수 (버터플라이 매도 put(변동성 커질때))
    put_sell_price = -round(
        df_put[str(left)][i] + df_put[str(right)][i] - (2 * df_put[str(center)][i]), 4
    )

    buy_data = pd.DataFrame(
        {"일자": date, "S1_CALL": call_buy_price, "S1_PUT": put_buy_price}, index=[0]
    )
    sell_data = pd.DataFrame(
        {"일자": date, "S1_CALL": call_sell_price, "S1_PUT": put_sell_price}, index=[0]
    )
    butterfly_buy_1 = pd.concat([butterfly_buy_1, buy_data])
    butterfly_sell_1 = pd.concat([butterfly_sell_1, buy_data])
    butterfly_buy_1 = butterfly_buy_1.reset_index(drop=True).fillna(0)
    butterfly_sell_1 = butterfly_sell_1.reset_index(drop=True).fillna(0)
```
