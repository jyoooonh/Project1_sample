#### 니케이 변동성 (7/31~10/20)
* 야간 변동율 = (당일시가-전날종가)/전날종가 night_change
* 일본 없는 날짜 :2023-08-11,2023-09-18 -> 야간변동율 0값으로 처리(엑셀전처리)

```python
import pandas as pd
import os
```

```python
niv = pd.read_csv("../dataset/jniv_day.csv")

niv.columns=['date','close','open','high','low','tran','change'] 
niv.drop(['high','low','tran','change'],axis=1, inplace=True)
niv= niv[['date','open','close']]
niv['date'] = pd.to_datetime(niv['date'])
```

```python
niv = niv.sort_values(by='date')
niv.reset_index(drop=True, inplace=True)
```

```python
niv.head(5)
```

```python
# 파생변수 생성(야간 변동율 = night_change)
niv['close_shift'] = niv['close'].shift(1)
niv['niv_night_change'] = round((niv['open']-niv['close_shift'])/niv['close_shift']*100,3)
niv.drop('close_shift',axis=1,inplace=True)
```

```python
niv = niv.loc[1:,]
niv = niv[['date','niv_night_change']]
```

```python
niv
```

```python
df = pd.read_csv("../dataset/usd_krw_final.csv",index_col=0)
df['date'] = pd.to_datetime(df['date'])
```

```python
niv_final = pd.merge(df,niv,on='date',how='inner')
```
